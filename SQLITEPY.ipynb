```python
## v pythonu takto 

import sqlite3

conection = sqlite3.connect("structures.db")

cursor = conection.cursor()



## dotaz do databaze 
cursor.execute("""
    SELECT id, name FROM structures
        WHERE name LIKE "%benzene%";
""")

## cursor se chová jako ITERATOR, přes množinu výsledků 
for id, name in cursor: 
    print("%-30s %s" % (name, id))

cursor.close()
conection.close()
```

```python
import sqlite3


conection = sqlite3.connect("pokus13.db")
cursor = conection.cursor()


cursor.execute("""
    DROP TABLE IF EXISTS pokus.db;
"""
)

cursor.execute ("""
    CREATE TABLE oddeleni (
               nazev TEXT, 
               cislo INTEGER 
);          
"""                 
)
cursor.execute ("""
    CREATE TABLE zamestnanci(
               prijmeni TEXT, 
               oddil INTEGER              
);
"""
)
cursor.execute("""
    INSERT INTO oddeleni (nazev, cislo) VALUES
    ("personal",1),
    ("mzdové",2),
    ("uklid", 3);
    """
)
cursor.execute("""
    INSERT INTO zamestnanci (prijmeni, oddil) VALUES
    ("novak",1),
    ("pepina",2),
    ("novak2", 3);            
"""
)

print(cursor)

cursor.close()
conection.close()

```
